```python
import numpy as np
import matplotlib.pyplot as plt
```

Example 1

```python
idx = np.linspace(0,6.28,num=100)
query = np.sin(idx) + np.random.uniform(size=100)/10.0
```

```python
template = np.cos(idx*2)
```

```python
from dtw import *
alignment = dtw(query, template, keep_internals=True)
```

```python
alignment.plot(type="threeway")

plt.show()
```

```python
dtw(query, template, keep_internals=True, 
    step_pattern=rabinerJuangStepPattern(6, "c"))\
    .plot(type="twoway",offset=-2)

plt.show()
```

```python
print(rabinerJuangStepPattern(6,"c"))
rabinerJuangStepPattern(6,"c").plot()
```

```python
#plotting the path of least cost on the cost matrix

alignment = dtw(query, template, keep_internals=True, 
    window_type="sakoechiba", window_args={'window_size': 10})\
    .plot()
```

Example 2

```python
#using some noisy trig functions

idx = np.linspace(0, 6.28, num=100)

x = np.sin(0.6*idx)

y = np.cos(idx) + np.random.uniform(size=100)

alignment = dtw(x, y, keep_internals=True)
```

```python
alignment.plot(type="threeway")
```

```python
dtw(y, x, keep_internals=True, 
    step_pattern=rabinerJuangStepPattern(6, "c"))\
    .plot(type="twoway", offset=-4)
```

```python
print(rabinerJuangStepPattern(6,"c"))
rabinerJuangStepPattern(6,"c").plot()

plt.show()
```

```python
#Cost Matrix with path of least cost plotted

alignment = dtw(x, y, keep_internals=True,
               window_type="sakoechiba", window_args={'window_size':10})\
            .plot()

```
